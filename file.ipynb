**TOOK 3 CLUSTERS FOR  GRAPH AND KMEAN**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv("1.csv")
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
 
features = data[['Age', 'Annual Income (k$)']]

kmeans = KMeans(n_clusters=3)
kmeans.fit(features)
cluster_labels = kmeans.labels_
plt.scatter(features['Age'], features['Annual Income (k$)'], c=cluster_labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('K-means Clustering')
plt.show()
```

**Elbow Method to get Optimal Cluster**

```python
sse = []
for k in range(1, 55):
    kmeans = KMeans(n_clusters=k, max_iter=300, random_state=0)
    kmeans.fit(features)
    sse.append(kmeans.inertia_)

number_clusters = range(1, 55)
plt.plot(number_clusters, sse)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
# Assuming the optimum number of clusters is 10 and you have new data 'new_data'
kmeans = KMeans(n_clusters=3, max_iter=300, n_init=10, random_state=0)
kmeans.fit(features)
labels = kmeans.predict(features)
labels
```

**Based On this Got 10 as optimal cluster and then done 10 cluster k-mean**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv("1.csv")
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
 
features = data[['Age', 'Annual Income (k$)']]



num_clusters = 20
kmeans = KMeans(n_clusters=num_clusters, max_iter=300, random_state=0)
kmeans.fit(features)

# Get cluster labels
cluster_labels = kmeans.labels_
data['Cluster'] = cluster_labels
# Visualize the clusters
plt.scatter(features['Age'], features['Annual Income (k$)'], c=cluster_labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('K-means Clustering with 10 clusters')
plt.show()
```
